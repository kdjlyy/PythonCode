```python
import matplotlib.pyplot as plt
import random
import matplotlib

# Windows、Linux设置中文字体方法
# font = {'family' : 'MicroSoft YaHei',
#         'weight' : 'bold',
#       }
# matplotlib.rc('font', **font)  # pass in the font dict as kwargs
# plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
x = range(0, 120)
y = [random.randint(20, 35) for i in range(120)]

plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)

# 调整x轴的刻度
_x = list(x)
_xtick_labels = ["{}:{:02}".format((10+i//60), i%60) for i in _x]

# 这里必须传入两个参数，让字符串和数字一一对应
plt.xticks(_x[::5], _xtick_labels[::5], rotation=-45) # rotation旋转度数

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度 单位(℃)")
plt.title("10点到12点每分钟的气温变化情况")

plt.show()
```
